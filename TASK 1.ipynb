**Data Cleaning**

```python
df_t.isnull().sum()
```

```python
# Create proper copy
df_t = df_t.copy()

# Fill missing Age values
df_t['Age'] = df_t['Age'].fillna(df_t['Age'].median())

# Fill missing Embarked values
df_t['Embarked'] = df_t['Embarked'].fillna(df_t['Embarked'].mode()[0])

# Remove duplicates
df_t = df_t.drop_duplicates()
# Check missing values
print(df_t.isnull().sum())
print(df_t.shape)
```

```python
df_t = df_t.drop('Cabin', axis=1)
```

**Visualisations**

```python
#HISTOGRAM
df_t['Age'].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#HISTOGRAM
df_t['Fare'].hist()
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
#BAR CHART
sns.countplot(x='Sex', data=df_t)
plt.title("Gender Count")
plt.show()
```

```python
#BAR CHART
sns.countplot(x='Embarked', data=df_t)
plt.title("Embarked Count")
plt.show()
```

```python
#BOX PLOT
plt.figure(figsize=(10,6))

sns.boxplot(data=df_t[['Age', 'Fare']])

plt.title("Box Plots of Age and Fare")

plt.show()
```

```python
#COUNT PLOT
plt.figure(figsize=(8,5))

sns.countplot(x='Survived', data=df_t)

plt.title("Survival Count")

plt.xlabel("Survived")
plt.ylabel("Count")

plt.show()
```

```python
#SCATTERPLOT
plt.figure(figsize=(8,5))

sns.scatterplot(x='Age',
                y='Fare',
                data=df_t)

plt.title("Age vs Fare")

plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

```python
#PAIR PLOT
sns.pairplot(df_t[['Age','Fare','Pclass','SibSp','Parch']])

plt.show()
```

```python
#CORRELATION HEATMAP USING SEABORN
plt.figure(figsize=(10,6))

sns.heatmap(df_t.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
#MISSING VALUES HEATMAP
plt.figure(figsize=(10,6))

sns.heatmap(df_t.isnull(),
            cbar=False,
            cmap='viridis')

plt.title("Missing Values Heatmap")

plt.show()
```

**KEY INSIGHTS FROM MY VISUALIZATIONS**

**1.** Most passengers were between the ages of 20 and 40, indicating that the dataset mainly consisted of young and middle-aged adults.

**2.** Female passengers had a significantly higher survival rate compared to male passengers, showing a strong relationship between gender and survival.

**3.** Passengers traveling in First Class paid much higher fares and also showed better survival chances than passengers in lower classes.

**4.** The Histogram of fares showed a right-skewed distribution, meaning a small number of passengers paid extremely high ticket prices.

**5.** The correlation heatmap revealed that passenger class and fare were strongly related, as higher-class passengers generally paid more.

**6.** Count plots showed that the majority of passengers were male, which created an imbalance in gender distribution within the dataset.

**Surprising Patterns and Anomalies From this dataset**

**1.** Women survived more than men, even though there were fewer women on the ship.

**2.** Some passengers paid very high ticket prices compared to others.

**3.** A few old passengers still survived, even though survival was generally lower for older people.

**4.** Some passengers with cheap tickets also survived, so money was not the only factor for survival.
